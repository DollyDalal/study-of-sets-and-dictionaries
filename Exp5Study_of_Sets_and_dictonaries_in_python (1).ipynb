```python
#problem statement 3: one set contains students in the cricket club, another in the football club, and one has in the badminton club.
#find students who are: 1. in both both clubs. 2. in only one club3. in all three clubs. 4. student in neither club

# Sets of students in clubs
cricket = {"aman", "karan", "priya", "rohan"}
football = {"karan", "priya", "dolly", "aayush"}
badminton = {"aman", "dolly", "shanvi"}

# Universal set (all students)
all_students = {"aman", "karan", "priya", "rohan", "dolly", "aayush", "shanvi", "trisha"}

# 1. Students in both cricket and football clubs (intersection)
both_clubs = cricket & football

# 2. Students in only one club (symmetric difference)
only_one_club = cricket ^ football ^ badminton

# 3. Students in all three clubs (intersection)
all_three_clubs = cricket & football & badminton

# 4. Students in neither club
neither_club = all_students - (cricket | football | badminton)

# Output
print("Students in both cricket and football clubs:", both_clubs)
print("Students in only one club:", only_one_club)
print("Students in all three clubs:", all_three_clubs)
print("Students in neither club:", neither_club)


```

```python
#problem statement 4: absent students
#given a set of all students and a set of present students, find the absent students.

# Set of all students
all_students = {"aman", "karan", "priya", "rohan", "dolly", "aayush", "shanvi"}
present_students = {"aman", "priya", "dolly", "shanvi","aayush"}
absent_students = all_students - present_students

print("Absent students are:")
print(absent_students)
```
